```python
import pyspark.pandas as ps  # Importa la API de pandas sobre Spark (permite usar sintaxis tipo pandas en entornos distribuidos de Spark)

import matplotlib.pyplot as plt  # Importa la librer칤a Matplotlib para crear gr치ficos y visualizaciones en 2D

import seaborn as sns  # Importa Seaborn, una librer칤a basada en Matplotlib que permite crear gr치ficos estad칤sticos m치s atractivos y f치ciles

import plotly.express as px  # Importa Plotly Express, una librer칤a para crear gr치ficos interactivos de manera sencilla


```

```python
# Inicializar Spark primero
from pyspark.sql import SparkSession
#Crea o recupera una sesi칩n activa de Spark.
spark = SparkSession.builder \
    .appName("Global Terrorismo") \
    .getOrCreate()

# Importar pandas API on Spark
import pyspark.pandas as ks  # Koalas



```

```python
tbl_name = 'default.global_terrorism'  # Define el nombre completo de la tabla que se leer치 desde el cat치logo de Spark 
                                       # ('default' es la base de datos y 'global_terrorism' la tabla)

df = spark.read.table(tbl_name)        # Usa Spark para leer la tabla indicada y cargarla en un DataFrame distribuido (df)

df_pd = df.toPandas()                  # Convierte el DataFrame de Spark en un DataFrame de pandas 
                                       # (칰til para trabajar localmente o visualizar los datos)

df_pd.head(5)                          # Muestra las primeras 5 filas del DataFrame en formato de tabla
```

**_OPERACIONES BASICAS DE ANALISIS DE DATOS._**

```python
#Estadisticas descriptivas
df_pd.describe()
```

```python
#Contar valores unicos en una columna
df_pd['country_txt'].nunique()
```

```python
df_2010 = df_pd[df_pd['iyear'] >= 2010]  
# Filtra el DataFrame original (df_pd) para quedarse solo con los registros
# donde el a침o del atentado ('iyear') sea mayor o igual a 2010.
# El resultado se guarda en un nuevo DataFrame llamado df_2010.


df_2010.groupby('region_txt')['suicide'].sum().sort_values(ascending=False)  
# Agrupa los datos por la columna 'region_txt' (nombre de la regi칩n)
# Luego suma los valores de la columna 'suicide' (cantidad de suicidios en cada regi칩n)
# Finalmente ordena los resultados de mayor a menor con 'ascending=False',
# mostrando las regiones con m치s suicidios primero.


```

```python
#N칰mero de ataques por a침o

df_pd['iyear'].value_counts().sort_index().plot(kind='bar',figsize=(15,5))
plt.title('N칰mero de ataques por a침o')
plt.xlabel('A침o')
plt.ylabel('Cantidad de ataques')
plt.show()
```

```python
sns.countplot(data=df_pd, x='attacktype1')  
# Crea un gr치fico de barras (countplot) usando Seaborn.  
# Muestra la cantidad de ocurrencias de cada tipo de ataque seg칰n la columna 'attacktype1' del DataFrame df_pd.


plt.xticks(rotation=45)  
# Rota las etiquetas del eje X 45 grados para que se vean mejor (evita que se encimen los textos).


plt.title('Cantidad de ataques por tipo')  
# Agrega un t칤tulo al gr치fico para describir lo que muestra.


plt.show()  
# Muestra el gr치fico en pantalla.

```
